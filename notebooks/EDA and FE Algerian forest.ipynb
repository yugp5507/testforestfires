```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

## Data cleaning ##

```python
#  missing values
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:123,"Region"]=0
dataset.loc[123:,"Region"]=1
df=dataset
```

```python
df.drop(index=122, inplace=True)
```

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
df.info()
```

```python

df[['Region']] = df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# remove the null values
df =  df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
# fix space in columns names
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df[['month','day','year','Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
df.head()
```

## Changing the other columns to float data datatype ##

```python
objects = [feature for feature in df.columns if df[feature].dtype == 'object']
```

```python
objects
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
objects
```

```python
df.describe()
```

```python
# Let ave the Cleaned dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```
